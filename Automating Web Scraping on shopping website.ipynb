```python
from autoscraper import AutoScraper
```

```python
flipkart_url="https://www.flipkart.com/laptops-store?otracker=nmenu_sub_Electronics_0_Laptops"
wanted_list=["₹55,990","ASUS TUF Gaming F15 Core i5 10th Gen - (8 GB/512 GB SSD/Windows 10 Home/4 GB Graphics/NVIDIA GeForce GTX 1650 Ti/60 Hz) FX506LI-BQ057T Gaming Laptop",'4.5',"(730)"]
```

```python
scraper=AutoScraper()
result=scraper.build(flipkart_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(flipkart_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_n1wi':'Title','rule_pfzo':'Price','rule_4e8l':'Stars'})
scraper.keep_rules(['rule_tvkc','rule_q6eq','rule_1jz4'])
scraper.save('Flipkart-search')
```

```python
results=scraper.get_result_similar('https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=3c383f9e-381c-44ee-88ba-d8f73f8c00cf&as-searchtext=lap',group_by_alias=True)
```
